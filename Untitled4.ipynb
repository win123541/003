```python
import pandas as pd 
import alphalens as al
from scipy.stats import stats
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestRegressor 
```

```python
financial =  pd.read_csv('financial.csv')
index_c = pd.read_csv('index_constituents.csv')
price = pd.read_csv('price.csv')
```

```python
val_fac_use= ['public_date','gvkey','ps','pcf','cash_lt','debt_ebitda','de_ratio','opmad']
financial = financial[val_fac_use]
financial
```

```python
def rolling_pred(train_feature, train_return, test_feature, test_return):
    '''
    This function suppose to use train_feature, train_return to generate a model and use test_feature to predict
    the test_returns. 
    Ouput:
        test_return
    '''
    rf = RandomForestRegressor(n_estimators = 10, random_state = 42)
    rf.fit(train_feature, train_return)
    test_predict = rf.predict(test_feature)
    errors = abs(test_predict - test_return)
    print('mean square error: ', round(np.mean(errors**2), 2), 'degrees')
    return test_predict
```

```python
price.columns = ['gvkey', 'date', 'close_price']
val_fac_use= ['date','gvkey','ps','pcf','cash_lt','debt_ebitda','de_ratio','opmad']
financial.columns = val_fac_use
price
```

```python
price_table = pd.pivot_table(price, values = 'close_price',index = ['date'],columns = 'gvkey',aggfunc=np.sum)
price_table2 = price_table.diff(-60).copy()
price_table = price_table2/price_table
price_table = price_table.stack()
price_table = price_table.reset_index()
price_table.columns = ['date','gvkey','return']
price_table
```

```python
financial = pd.pivot_table(financial, index = ['date'], columns = 'gvkey', aggfunc = np.mean)
financial = financial.stack()
financial
```

```python
result = pd.merge(financial, price_table, 'right', on = ['date','gvkey'])
```

```python
result = result.dropna()
result
```

```python
qtr_date = [20160000,20160400,20160700,20161000,20170000,20170400,20170700,20171000,20180000,
           20180400,20180700,20181000,20190000,20190400,20190700,20191000,20200000]
feature_list = ['cash_lt', 'de_ratio', 'debt_ebitda', 'opmad', 'pcf', 'ps']
```

```python
predictions = np.array([])
for i in range(len(qtr_date)-5):
    window = result[qtr_date[i] <= result['date']]
    window = window[qtr_date[i+5] >= window['date']]
    train = window[qtr_date[i+4]>= window['date']]
    train_f = train[feature_list]
    train_r = train['return']
    test = window[qtr_date[i+4]<= window['date']]
    test_f = test[feature_list]
    test_r = test['return']
    predictions = np.concatenate((predictions, rolling_pred(train_f, train_r, test_f, test_r)))
```

```python
result = result[20170000<=result['date']]
result['pred'] = predictions
result #ready to put in alphalens
```
