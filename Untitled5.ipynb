```python
price2['date'] =price.index+pd.DateOffset(months = 3)
price2 = price2.set_index('date')
price2.columns = ['gvkey','3MB_close']
price2
```

```python
price_new = pd.merge(price,price2, 'right', on = ['date','gvkey'])
price_new
```

```python
price_new['return'] = (price_new['close_price']-price_new['3MB_close'])/price_new['3MB_close']
price_new = price_new.fillna(method = 'ffill')
price_new = price_new.drop('close_price', axis = 1)
price_new = price_new.drop('3MB_close', axis = 1)
price_new.index = price_new.index-pd.DateOffset(months = 3)
price_new
```

```python
result = pd.merge(financial, price_new,'right', on = ['date','gvkey'])
result = result.dropna()
result
```

```python
begin_date = datetime.datetime(2016,1,1)
end_date = datetime.datetime(2020,1,5)
qtr_list = pd.date_range(begin_date, end_date, freq="BQS")
feature_list = ['cash_lt', 'de_ratio', 'debt_ebitda', 'opmad', 'pcf', 'ps']
qtr_list
```

```python
def rolling_pred(train_feature, train_return, test_feature, test_return):
    '''
    This function suppose to use train_feature, train_return to generate a model and use test_feature to predict
    the test_returns. 
    Ouput:
        test_return
    '''
    rf = RandomForestRegressor(n_estimators = 10, random_state = 42)
    rf.fit(train_feature, train_return)
    test_predict = rf.predict(test_feature)
    errors = abs(test_predict - test_return)
    print('mean square error: ', round(np.mean(errors**2), 5), 'degrees')
    return test_predict
```

```python
predictions = np.array([])
for i in range(len(qtr_list)-6):
    train = result[qtr_list[i]:qtr_list[i+5]]
    train_f = train[feature_list]
    train_r = train['return']
    test = result[qtr_list[i+5]:qtr_list[i+6]]
    test_f = test[feature_list]
    test_r = test['return']
    predictions = np.concatenate((predictions, rolling_pred(train_f, train_r, test_f, test_r)))
```

```python
result = result[qtr_list[5]:]
result['pred'] = predictions
result #ready to put in alphalens
```
